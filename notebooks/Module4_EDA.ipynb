```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Libraries Imported Successfully")
```

```python
ap_df = pd.read_csv("../dataset/processed/clean_antennas_dataset.csv")

es_df = pd.read_csv("../dataset/processed/clean_embedded_dataset.csv")

print("Datasets Loaded Successfully")
```

```python
ap_df["Priority"].value_counts()
```

```python
es_df["Priority"].value_counts()
```

```python
ap_df["Priority"].value_counts().plot(kind="bar")

plt.title("Priority Distribution")

plt.xlabel("Priority")

plt.ylabel("Number of Questions")

plt.show()
```

```python
ap_df["Chapter"].value_counts()
```

```python
ap_df["Chapter"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Chapters")

plt.xlabel("Chapter")

plt.ylabel("Questions")

plt.show()
```

```python
ap_df["Difficulty"].value_counts().plot(kind="pie",autopct="%1.1f%%")

plt.title("Difficulty Distribution")

plt.show()
```

```python
ap_df["Average_Marks"].hist()

plt.title("Marks Distribution")

plt.xlabel("Marks")

plt.ylabel("Frequency")

plt.show()
```

```python
ap_df["Years_Appeared"].value_counts().sort_index().plot()

plt.title("Year Wise Question Trend")

plt.xlabel("Year")

plt.ylabel("Questions")

plt.show()
```
